```
from IPython.display import display, clear_output
import ipywidgets as widgets
import random
longitud = 4
modo_maquina = False
def generar_numero(longitud=4):
    digitos = list('0123456789')
    random.shuffle(digitos)
    return ''.join(digitos[:longitud])
def evaluar(numero_secreto, intento):
    fijas = sum(1 for i in range(len(intento)) if intento[i] == numero_secreto[i])
    picas = sum(1 for i in range(len(intento)) if intento[i] in numero_secreto and intento[i] != numero_secreto[i])
    return picas, fijas
numero_secreto = generar_numero(longitud)
intento_input = widgets.Text(value='', placeholder='Escribe tu intento', description='Intento:')
resultado_output = widgets.Output()
boton = widgets.Button(description='Evaluar', button_style='success')
reiniciar_btn = widgets.Button(description='Reiniciar', button_style='warning')
historial = []
def evaluar_intento(b):
    intento = intento_input.value.strip()
    if len(intento) != longitud or not intento.isdigit():
        with resultado_output:
            clear_output()
            print(f"Ingresa un número de {longitud} dígitos válidos.")
        return
    picas, fijas = evaluar(numero_secreto, intento)
    historial.append((intento, picas, fijas))
    with resultado_output:
        clear_output()
        for h in historial:
            print(f"Intento {h[0]} → {h[1]} picas, {h[2]} fijas")
        if fijas == longitud:
            print(f"Lo admito, ganaste, El número era {numero_secreto}.")
            boton.disabled = True
def reiniciar_juego(b):
    global numero_secreto, historial
    numero_secreto = generar_numero(longitud)
    historial = []
    intento_input.value = ''
    boton.disabled = False
    with resultado_output:
        clear_output()
        print("Empezamos. Adivina el número")
boton.on_click(evaluar_intento)
reiniciar_btn.on_click(reiniciar_juego)
print(f"Se ha generado un número secreto de {longitud} dígitos únicos.")
display(intento_input, boton, reiniciar_btn, resultado_output)
def maquina_adivina(secreto):
    print("intentare adivinar tu número")
    posibles = [''.join(p) for p in __import__('itertools').permutations('0123456789', len(secreto))]
    intentos = 0
    while posibles:
        intento = random.choice(posibles)
        intentos += 1
        p, f = evaluar(secreto, intento)
        print(f"Intento {intentos}: {intento} → {p} picas, {f} fijas")
        if f == len(secreto):
            print(f"adivine el número {secreto} en {intentos} intentos")
            break
        posibles = [x for x in posibles if evaluar(intento, x) == (p, f)]
```
