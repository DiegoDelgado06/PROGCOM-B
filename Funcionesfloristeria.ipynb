```python
inventario = {
    "rosas": {"precio": 3500, "stock": 50},
    "girasoles": {"precio": 4000, "stock": 30},
    "lirios": {"precio": 5000, "stock": 20},
    "margaritas": {"precio": 2800, "stock": 40},
    "orquideas": {"precio": 8500, "stock": 10},
}
ventas = []
def presentar_floristeria():
    print("\nBienvenido a la Floriteria DADH")
    print("Una floristería de ladrillo con techo gris,")
    print("plantas trepadoras en la fachada y una ventana con una persona asomada.")
    print("Un grafiti con las letras DADH adorna la parte trasera del local.\n")
def ver_inventario():
    print("Inventario disponible:")
    for flor, datos in inventario.items():
        if datos["stock"] > 0:
            print(f"{flor.capitalize()}: {datos['stock']} unidades | Precio: ${datos['precio']} COP")
    print("")
def registrar_venta(nombre, cantidad):
    nombre = nombre.lower()
    if nombre not in inventario:
        print(f"Error: La flor '{nombre}' no está disponible.")
        return False
    if inventario[nombre]["stock"] == 0:
        print(f"Error: La flor '{nombre}' está agotada.")
        return False
    if cantidad > inventario[nombre]["stock"]:
        print(f"Error: Solo quedan {inventario[nombre]['stock']} unidades de '{nombre}'.")
        return False
    total = cantidad * inventario[nombre]["precio"]
    inventario[nombre]["stock"] -= cantidad
    ventas.append({
        "flor": nombre,
        "cantidad": cantidad,
        "total": total
    })
    print(f"Venta exitosa: {cantidad} x {nombre.capitalize()} = ${total} COP\n")
    return True
def ver_ventas():
    if not ventas:
        print("No hay ventas registradas.\n")
    else:
        print("Historial de ventas:")
        for venta in ventas:
            print(f"{venta['cantidad']} x {venta['flor'].capitalize()} = ${venta['total']} COP")
        print("")
def ingresos_totales():
    total = sum(v["total"] for v in ventas)
    print(f"Ingresos totales: ${total} COP\n")
def menu_cliente():
    presentar_floristeria()
    while True:
        print("Opciones:")
        print("1. Ver inventario")
        print("2. Comprar flor")
        print("3. Ver historial de ventas")
        print("4. Ver ingresos totales")
        print("5. Salir")
        opcion = input("Seleccione una opción (1-5): ")
        if opcion == "1":
            ver_inventario()
        elif opcion == "2":
            nombre = input("Ingrese el nombre de la flor a comprar: ").lower()
            try:
                cantidad = int(input("Cantidad: "))
                if cantidad <= 0:
                    print("Error: La cantidad debe ser mayor que cero.\n")
                    continue
                registrar_venta(nombre, cantidad)
            except:
                print("Error: Cantidad inválida.\n")
        elif opcion == "3":
            ver_ventas()
        elif opcion == "4":
            ingresos_totales()
        elif opcion == "5":
            print("Gracias por visitar Floriteria DADH. Hasta luego.")
            break
        else:
            print("Opción inválida.\n")
menu_cliente()
```
