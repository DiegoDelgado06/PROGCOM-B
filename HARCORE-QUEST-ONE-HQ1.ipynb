```python
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

# ParÃ¡metros del juego
TAMANO_TABLERO = 8
NUMERO_MINAS = 10

# Variables globales
tablero = []
botones = []
turno = 1

# Etiqueta de turno
label_turno = widgets.Label()

# Contenedor de botones
button_grid = widgets.GridBox(layout=widgets.Layout(grid_template_columns="repeat(8, 50px)", grid_gap="2px"))

def generar_tablero():
    tablero = [[0 for _ in range(TAMANO_TABLERO)] for _ in range(TAMANO_TABLERO)]
    minas_colocadas = 0
    while minas_colocadas < NUMERO_MINAS:
        fila = random.randint(0, TAMANO_TABLERO - 1)
        columna = random.randint(0, TAMANO_TABLERO - 1)
        if tablero[fila][columna] != -1:
            tablero[fila][columna] = -1
            minas_colocadas += 1
    for i in range(TAMANO_TABLERO):
        for j in range(TAMANO_TABLERO):
            if tablero[i][j] == -1:
                continue
            minas = 0
            for di in range(-1, 2):
                for dj in range(-1, 2):
                    ni, nj = i + di, j + dj
                    if 0 <= ni < TAMANO_TABLERO and 0 <= nj < TAMANO_TABLERO:
                        if tablero[ni][nj] == -1:
                            minas += 1
            tablero[i][j] = minas
    return tablero

def mostrar_mensaje(msg):
    clear_output(wait=True)
    display(label_turno, button_grid)
    print(msg)

def revelar_celda(i, j):
    global turno
    boton = botones[i][j]
    if boton.disabled:
        return
    if tablero[i][j] == -1:
        boton.description = "ðŸ’£"
        boton.style.button_color = 'red'
        deshabilitar_todo()
        mostrar_mensaje(f"ðŸ’¥ Jugador {turno} perdiÃ³. Â¡Fin del juego!")
        return
    elif tablero[i][j] == 0:
        expandir_ceros(i, j)
    else:
        boton.description = str(tablero[i][j])
    boton.disabled = True
    turno = 2 if turno == 1 else 1
    label_turno.value = f"Turno del Jugador {turno}"

def expandir_ceros(i, j):
    stack = [(i, j)]
    while stack:
        x, y = stack.pop()
        if not (0 <= x < TAMANO_TABLERO and 0 <= y < TAMANO_TABLERO):
            continue
        boton = botones[x][y]
        if boton.disabled:
            continue
        boton.disabled = True
        valor = tablero[x][y]
        boton.description = '' if valor == 0 else str(valor)
        if valor == 0:
            for dx in range(-1, 2):
                for dy in range(-1, 2):
                    nx, ny = x + dx, y + dy
                    stack.append((nx, ny))

def deshabilitar_todo():
    for fila in botones:
        for b in fila:
            b.disabled = True
            if tablero[botones.index(fila)][fila.index(b)] == -1 and b.description != "ðŸ’£":
                b.description = "ðŸ’£"

def crear_tablero():
    global tablero, botones, turno
    tablero = generar_tablero()
    botones = []
    turno = 1
    label_turno.value = "Turno del Jugador 1"
    button_widgets = []
    for i in range(TAMANO_TABLERO):
        fila = []
        for j in range(TAMANO_TABLERO):
            btn = widgets.Button(
                description='',
                layout=widgets.Layout(width='50px', height='50px'),
                style=widgets.ButtonStyle()
            )
            btn.on_click(lambda b, i=i, j=j: revelar_celda(i, j))
            fila.append(btn)
            button_widgets.append(btn)
        botones.append(fila)
    button_grid.children = button_widgets

# Iniciar juego
crear_tablero()
display(label_turno, button_grid)
```
