```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatText, IntText, Dropdown, Button, VBox, HBox, Output
from IPython.display import display
from google.colab import files
def generar_tabla_amortizacion(monto, tasa_anual, plazo, frecuencia):
    """
    Genera la tabla de amortización para un préstamo.
    """
    if frecuencia == "Mensual":
        n = 12
    elif frecuencia == "Trimestral":
        n = 4
    elif frecuencia == "Semestral":
        n = 2
    else:
        n = 1
    tasa_periodica = tasa_anual / 100 / n
    num_pagos = plazo * n
    cuota = monto * (tasa_periodica * (1 + tasa_periodica) ** num_pagos) / ((1 + tasa_periodica) ** num_pagos - 1)
    saldo = monto
    datos = []
    for i in range(1, num_pagos + 1):
        interes = saldo * tasa_periodica
        capital = cuota - interes
        saldo -= capital
        saldo = max(0, saldo)
        datos.append([i, round(cuota, 2), round(interes, 2), round(capital, 2), round(saldo, 2)])
    df = pd.DataFrame(datos, columns=["Periodo", "Cuota", "Interés", "Capital", "Saldo"])
    return df, cuota
def graficar_amortizacion(df):
    plt.figure(figsize=(10, 6))
    plt.plot(df["Periodo"], df["Saldo"], label="Saldo Restante", linewidth=2)
    plt.plot(df["Periodo"], df["Interés"], label="Interés Pagado", linestyle="--")
    plt.plot(df["Periodo"], df["Capital"], label="Capital Amortizado", linestyle=":")
    plt.title("Evolución del Préstamo", fontsize=14)
    plt.xlabel("Periodo")
    plt.ylabel("Valor ($)")
    plt.legend()
    plt.grid(True)
    plt.show()
monto_input = FloatText(value=10000000, description="Monto ($):", step=100000)
tasa_input = FloatText(value=12.0, description="Tasa Anual (%):")
plazo_input = IntText(value=5, description="Plazo (años):")
frecuencia_input = Dropdown(options=["Mensual", "Trimestral", "Semestral", "Anual"], value="Mensual", description="Frecuencia:")
boton_generar = Button(description="Generar Tabla", button_style="success")
boton_exportar = Button(description="Exportar a Excel", button_style="info")
salida = Output()
def ejecutar(_):
    with salida:
        salida.clear_output()
        monto = monto_input.value
        tasa = tasa_input.value
        plazo = plazo_input.value
        frecuencia = frecuencia_input.value
        df, cuota = generar_tabla_amortizacion(monto, tasa, plazo, frecuencia)
        print(f"Cuota fija: ${cuota:,.2f}")
        display(df)
        graficar_amortizacion(df)
def exportar_excel(_):
    with salida:
        salida.clear_output()
        monto = monto_input.value
        tasa = tasa_input.value
        plazo = plazo_input.value
        frecuencia = frecuencia_input.value
        df, cuota = generar_tabla_amortizacion(monto, tasa, plazo, frecuencia)
        nombre = "tabla_amortizacion.xlsx"
        df.to_excel(nombre, index=False)
        print(f"Archivo exportado correctamente como '{nombre}'")
        files.download(nombre)
boton_generar.on_click(ejecutar)
boton_exportar.on_click(exportar_excel)
ui = VBox([
    HBox([monto_input, tasa_input]),
    HBox([plazo_input, frecuencia_input]),
    HBox([boton_generar, boton_exportar]),
    salida
])
display(ui)
```
