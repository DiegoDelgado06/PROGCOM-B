```
import re
import ipywidgets as widgets
from IPython.display import display, clear_output
condiciones = [
    ("Debe tener al menos 4 caracteres.", lambda c: len(c) >= 4),
    ("Debe contener al menos un n√∫mero.", lambda c: any(ch.isdigit() for ch in c)),
    ("Debe contener al menos una letra may√∫scula.", lambda c: any(ch.isupper() for ch in c)),
    ("Debe contener al menos una letra min√∫scula.", lambda c: any(ch.islower() for ch in c)),
    ("Debe contener al menos un s√≠mbolo (!, @, #, $, %, &, etc.).", lambda c: bool(re.search(r"[!@#$%^&*(),.?\":{}|<>]", c))),
    ("Debe tener m√°s de 8 caracteres.", lambda c: len(c) > 8),
    ("No puede tener espacios.", lambda c: " " not in c),
    ("Debe comenzar con una letra.", lambda c: c[0].isalpha() if c else False),
    ("Debe incluir al menos dos n√∫meros.", lambda c: sum(ch.isdigit() for ch in c) >= 2),
    ("Debe tener una combinaci√≥n de may√∫sculas, min√∫sculas, n√∫meros y s√≠mbolos.",
     lambda c: all([
         any(ch.isupper() for ch in c),
         any(ch.islower() for ch in c),
         any(ch.isdigit() for ch in c),
         bool(re.search(r"[!@#$%^&*(),.?\":{}|<>]", c))
     ]))
]
nivel = 0
total = len(condiciones)
ultima_contrase√±a = ""
titulo = widgets.HTML("<h2>Juego de la Contrase√±a - Versi√≥n Acumulativa</h2>")
descripcion = widgets.HTML("<p>Cada nivel agrega una nueva condici√≥n que tu contrase√±a debe cumplir. No puedes romper las anteriores.</p>")
cond_label = widgets.HTML(f"<h4>Condici√≥n 1Ô∏è: {condiciones[0][0]}</h4>")
entrada = widgets.Password(placeholder='Escribe tu contrase√±a...')
boton = widgets.Button(description='Comprobar', button_style='info')
salida = widgets.Output()
def comprobar(b):
    global nivel, ultima_contrase√±a
    with salida:
        clear_output()
        contr = entrada.value
        cumple_todas = all(condiciones[i][1](contr) for i in range(nivel+1))
        if cumple_todas:
            print(f"Cumples las {nivel+1} condiciones actuales.")
            ultima_contrase√±a = contr
            nivel += 1
            if nivel < total:
                cond_label.value = f"<h4>Condici√≥n {nivel+1}Ô∏è‚É£: {condiciones[nivel][0]}</h4>"
                print("Ahora tu nueva contrase√±a debe seguir cumpliendo todas las condiciones anteriores m√°s esta nueva.\n")
            else:
                clear_output()
                display(widgets.HTML(f"<h2> Has completado las 10 condiciones bro, un 5 </h2>"))
                display(widgets.HTML(f"<b>Tu contrase√±a final fue:</b> <code>{ultima_contrase√±a}</code>"))
                boton.disabled = True
                entrada.disabled = True
        else:
            print("No cumples con todas las condiciones acumuladas. Intenta ajustar tu contrase√±a.\n")
            for i in range(nivel+1):
                estado = "‚úî" if condiciones[i][1](contr) else "‚úñ"
                print(f"{estado} {condiciones[i][0]}")
boton.on_click(comprobar)
display(titulo, descripcion, cond_label, entrada, boton, salida)
```
