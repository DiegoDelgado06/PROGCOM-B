```python
import random

# Opciones extendidas
opciones_extendido = [
    "Piedra", "Papel", "Tijera", "Lagarto", "Spock",
    "Pistola", "Ninja", "Pirata", "Meteorito", "Carro",
    "Drag√≥n", "Unicornio"
]

# Diccionario de reglas
reglas = {
    "Piedra":     ["Tijera", "Lagarto", "Pistola", "Meteorito", "Carro", "Drag√≥n", "Unicornio"],
    "Papel":      ["Piedra", "Spock", "Pistola", "Meteorito", "Carro", "Drag√≥n", "Unicornio"],
    "Tijera":     ["Papel", "Lagarto", "Pistola", "Ninja", "Carro", "Drag√≥n", "Unicornio"],
    "Lagarto":    ["Spock", "Papel", "Pistola", "Pirata", "Carro", "Drag√≥n", "Unicornio"],
    "Spock":      ["Tijera", "Piedra", "Pistola", "Ninja", "Carro", "Drag√≥n", "Unicornio"],
    "Pistola":    ["Spock", "Lagarto", "Tijera", "Pirata", "Carro", "Drag√≥n", "Unicornio"],
    "Ninja":      ["Pirata", "Papel", "Lagarto", "Pistola", "Carro", "Drag√≥n", "Unicornio"],
    "Pirata":     ["Papel", "Spock", "Meteorito", "Ninja", "Carro", "Drag√≥n", "Unicornio"],
    "Meteorito":  ["Piedra", "Tijera", "Lagarto", "Spock", "Carro", "Drag√≥n", "Unicornio"],
    "Carro":      ["Piedra", "Papel", "Tijera", "Lagarto", "Spock", "Pistola", "Ninja", "Pirata", "Meteorito", "Drag√≥n", "Unicornio"],
    "Drag√≥n":     ["Piedra", "Papel", "Tijera", "Lagarto", "Spock", "Pistola", "Ninja", "Pirata", "Meteorito", "Carro"],
    "Unicornio":  ["Piedra", "Papel", "Tijera", "Lagarto", "Spock", "Pistola", "Ninja", "Pirata", "Meteorito", "Carro"]
}

# Marcadores
ganados = 0
perdidos = 0
empates = 0

print("¬°Bienvenido a Piedra, Papel, Tijera EXTREMO! ü™®üìÑ‚úÇÔ∏èü¶éüññ + m√°s...\n")

while True:
    print("\nOpciones:")
    for i, opcion in enumerate(opciones_extendido, 1):
        print(f"{i}. {opcion}")

    try:
        eleccion_usuario = int(input("\nElige una opci√≥n (1-12) o 0 para salir: "))
        if eleccion_usuario == 0:
            break
        if eleccion_usuario < 1 or eleccion_usuario > len(opciones_extendido):
            print("Opci√≥n no v√°lida. Intenta de nuevo.")
            continue
    except ValueError:
        print("Entrada inv√°lida. Debe ser un n√∫mero.")
        continue

    user = opciones_extendido[eleccion_usuario - 1]
    bot = random.choice(opciones_extendido)

    print(f"\nT√∫ eliges: {user}")
    print(f"El bot elige: {bot}")

    if user == bot:
        resultado = "Empate"
        empates += 1
    elif bot in reglas[user]:
        resultado = "¬°Ganaste!"
        ganados += 1
    else:
        resultado = "Perdiste"
        perdidos += 1

    print(f"\nResultado: {resultado}")
    print(f"Marcador: ‚úÖ {ganados} | ‚ùå {perdidos} | ü§ù {empates}")

print("\nGracias por jugar üéâ")
```
