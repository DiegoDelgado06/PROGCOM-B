```
import math
def imprimir_tablero(tab):
    print()
    for i in range(3):
        row = " | ".join(tab[3*i:3*i+3])
        print(" " + row)
        if i < 2:
            print("---+---+---")
    print()
def ganador(tab):
    lines = [
        (0,1,2),(3,4,5),(6,7,8),
        (0,3,6),(1,4,7),(2,5,8),
        (0,4,8),(2,4,6)
    ]
    for a,b,c in lines:
        if tab[a] == tab[b] == tab[c] and tab[a] != " ":
            return tab[a]
    return None
def tablero_lleno(tab):
    return all(c != " " for c in tab)
def movimientos_validos(tab):
    return [i for i, c in enumerate(tab) if c == " "]
def minimax(tab, juez, jugador_max, jugador_min):
    g = ganador(tab)
    if g == jugador_max:
        return 1, None
    elif g == jugador_min:
        return -1, None
    elif tablero_lleno(tab):
        return 0, None
    if juez == jugador_max:
        mejor_punt = -math.inf
        mejor_mov = None
        for mv in movimientos_validos(tab):
            tab[mv] = juez
            punt, _ = minimax(tab, jugador_min, jugador_max, jugador_min)
            tab[mv] = " "
            if punt > mejor_punt:
                mejor_punt = punt
                mejor_mov = mv
                if mejor_punt == 1:
                    break
        return mejor_punt, mejor_mov
    else:
        peor_punt = math.inf
        peor_mov = None
        for mv in movimientos_validos(tab):
            tab[mv] = juez
            punt, _ = minimax(tab, jugador_max, jugador_max, jugador_min)
            tab[mv] = " "
            if punt < peor_punt:
                peor_punt = punt
                peor_mov = mv
                if peor_punt == -1:
                    break
        return peor_punt, peor_mov
def movimiento_maquina(tab, maquina, jugador):
    _, mv = minimax(tab, maquina, maquina, jugador)
    return mv
def pedir_eleccion_simbolo():
    while True:
        elec = input("Elige el símbolo que vas a usar, rey (X/O): ").strip().upper()
        if elec in ("X","O"):
            return elec
        print("No escribiste bien. Escribe X o O.")
def pedir_orden_turnos(simbolo_jugador):
    if simbolo_jugador == "X":
        while True:
            r = input("¿Quieres ir tú de primero? (S/N): ").strip().upper()
            if r in ("S","N"):
                return r == "S"
            print("Responde S o N.")
    else:
        print("Como elegiste O, X comienza. La máquina jugará primero.")
        return False
def pedir_movimiento_usuario(tab):
    valid = movimientos_validos(tab)
    prompt = ("Es tu turno. Elige una casilla (1-9) según el siguiente mapa:\n"
              "1 2 3\n4 5 6\n7 8 9\nIngresa número: ")
    while True:
        try:
            v = input(prompt).strip()
            mv = int(v) - 1
            if mv in valid:
                return mv
            else:
                print("Casilla inválida o ocupada. Intenta otra, por favor.")
        except Exception:
            print("Entrada inválida. Ingresa un número del 1 al 9 correspondiente a una casilla libre, por favor.")
def jugar():
    print("Tic-Tac-Toe vs Máquina")
    jugador = pedir_eleccion_simbolo()
    maquina = "O" if jugador == "X" else "X"
    jugador_empieza = pedir_orden_turnos(jugador)
    tablero = [" "] * 9
    turno = "jugador" if jugador_empieza else "maquina"
    imprimir_tablero(tablero)
    while True:
        if turno == "jugador":
            mv = pedir_movimiento_usuario(tablero)
            tablero[mv] = jugador
            imprimir_tablero(tablero)
            g = ganador(tablero)
            if g:
                print(f"Resultado: {g} gana")
                break
            elif tablero_lleno(tablero):
                print("Fue empate, bro.")
                break
            turno = "maquina"
        else:
            print("Espérate, la máquina está pensando...")
            mv = movimiento_maquina(tablero, maquina, jugador)
            if mv is None:
                mv = movimientos_validos(tablero)[0]
            tablero[mv] = maquina
            imprimir_tablero(tablero)
            g = ganador(tablero)
            if g:
                print(f"Resultado: {g} gana")
                break
            elif tablero_lleno(tablero):
                print("Fue empate.")
                break
            turno = "jugador"
    print("Se acabó el juego.")
if __name__ == "__main__":
    jugar()
```
