```python
# C√≥digo para juego de Piedra, Papel, Tijera y sus 15 generalizaciones con interfaz gr√°fica did√°ctica en Google Colab

import random
from ipywidgets import widgets, VBox, Output, HTML
from IPython.display import display, clear_output

# Lista de los 15 elementos en orden circular
ELEMENTOS_15 = [
    "piedra", "fuego", "tijera", "serpiente", "humano",
    "√°rbol", "lobo", "esponja", "aire", "agua",
    "drag√≥n", "rayo", "pistola", "diablo", "papel"
]

# Crear diccionario: cada elemento vence a los 7 siguientes en el c√≠rculo
vence_a = {}
n = len(ELEMENTOS_15)
for i, elem in enumerate(ELEMENTOS_15):
    vence = [ELEMENTOS_15[(i + j + 1) % n] for j in range(7)]
    vence_a[elem] = vence

# Funci√≥n para determinar el resultado
def determinar_ganador(jugador, maquina):
    if jugador == maquina:
        return "Empate"  # Esto se evita en el c√≥digo
    elif maquina in vence_a[jugador]:
        return "¬°Ganaste! üéâ"
    else:
        return "Perdiste üò¢"

# Crear los widgets
dropdown = widgets.Dropdown(
    options=ELEMENTOS_15,
    description='Tu elecci√≥n:',
    layout=widgets.Layout(width='50%')
)

boton_jugar = widgets.Button(
    description='Jugar',
    button_style='success'
)

salida = Output()

instrucciones = HTML(
    value="""
    <h3>üïπÔ∏è Piedra, Papel, Tijera... y sus 15 generalizaciones</h3>
    <p><b>Instrucciones:</b><br>
    1. Elige un elemento del men√∫.<br>
    2. La m√°quina elegir√° otro aleatoriamente (sin empate).<br>
    3. Ganas si tu opci√≥n vence a la de la m√°quina (seg√∫n el ciclo circular de 15).<br>
    Cada elemento vence a los 7 siguientes en el c√≠rculo.<br>
    ¬°Divi√©rtete!</p>
    """
)

# Acci√≥n al hacer clic
def al_hacer_click(b):
    with salida:
        clear_output()
        jugador = dropdown.value
        maquina = random.choice([e for e in ELEMENTOS_15 if e != jugador])  # sin empate
        resultado = determinar_ganador(jugador, maquina)
        print(f"üßç T√∫ elegiste: {jugador}")
        print(f"ü§ñ La m√°quina eligi√≥: {maquina}")
        print(f"üèÅ Resultado: {resultado}")

boton_jugar.on_click(al_hacer_click)

# Mostrar interfaz
ui = VBox([instrucciones, dropdown, boton_jugar, salida])
display(ui)
```
